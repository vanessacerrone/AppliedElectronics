```python
import numpy as np
import pandas as pd
import scipy as sp
import scipy.optimize as op
import matplotlib.pyplot as plt
from matplotlib import gridspec
from matplotlib.ticker import AutoMinorLocator, MultipleLocator
from scipy.interpolate import UnivariateSpline
import math
%matplotlib inline
```

```python
# define ticks
def set_ticks(ax, xMaj, yMaj):
    ax.xaxis.set_major_locator(MultipleLocator(xMaj))
    ax.yaxis.set_major_locator(MultipleLocator(yMaj))
    ax.minorticks_on()
    ax.tick_params(which='major', width=1.0, length=8, direction='in', labelsize=12)
    ax.tick_params(which='minor', width=1.0, length=4, direction='in', labelsize=12)
```

```python
t,Vs1,Vs2,Vs1b,Vs2b,Vs1c,Vs2c = np.loadtxt('simulation.txt', unpack=True,skiprows=1)
```

```python
fig, ax = plt.subplots(figsize=(10,10))


ax.plot(t*1e09, Vs1, color='firebrick', ls='-', alpha=1, lw=2, label = 'V$_{\mathrm{s1}}$')
ax.plot(t*1e09, Vs2, color='lightcoral', ls='--', alpha=1, lw=2,label = 'V$_{\mathrm{s2}}$')

ax.plot(t*1e09, Vs1b, color='navy', ls='-', alpha=1, lw=2, label = 'V$_{\mathrm{s1b}}$')
ax.plot(t*1e09, Vs2b, color='royalblue', ls='--', alpha=1, lw=2,label = 'V$_{\mathrm{s2b}}$')

ax.plot(t*1e09, Vs1c, color='forestgreen', ls='-', alpha=1, lw=2, label = 'V$_{\mathrm{s1c}}$')
ax.plot(t*1e09, Vs2c, color='darkseagreen', ls='--', alpha=1, lw=2,label = 'V$_{\mathrm{s2c}}$')

# labels
ax.set_ylabel('Time [ns]', fontsize=14)
ax.set_xlabel('Voltage [V]', fontsize=14)

# ticks 
set_ticks(ax, 0.5,0.25)
ax.set_xlim(4,8)

ax.legend(loc=0, fontsize = 14, frameon=False,ncol=2)

plt.show()
```
